## Задача 1
Перестановка — это последовательность длины n целых чисел от 1 до n, в которой все числа встречаются ровно по одному разу. Например, (1), (4, 3, 5, 1, 2), (3, 2, 1) — перестановки, а (1, 1), (4, 3, 1), (2, 3, 4) — нет.

Бывает много различных задач на перестановки. Сегодня вам предстоит решить такую. Представьте, что Некто взял несколько перестановок (возможно, с разным количеством элементов), записал их в одну последовательность одну за другой, а затем перемешал полученную последовательность. Требуется восстановить исходные перестановки, если это возможно.

Входные данные
В первой строке содержится целое число n (1 ≤ n ≤ 105). Следующая строка содержит перемешанную последовательность из n целых чисел, разделенных одиночными пробелами. Числа в последовательности от 1 до 105.

Выходные данные
Если данную последовательность можно разбить на несколько перестановок так, что каждый элемент последовательности принадлежит в точности одной перестановке, в первой строке выведите количество получившихся перестановок. Вторая строка должна содержать n чисел, соответствующих элементам заданной последовательности. Если i-й элемент относится к первой перестановке, то i-е число должно быть 1, если ко второй — то 2, и т.д. Порядок нумерации перестановок произволен.

Если решений несколько, выведите любое. Если решения не существует, выведите в первой строке  - 1.
Примеры\
входные данные\
9\
1 2 3 1 2 1 4 2 5\
выходные данные\
3\
3 1 2 1 2 2 2 3 2

```python
def permutation(arr: list[int]):
    max_elem = max(arr)
    list_permutations = [0]*(max_elem+1)
    for i in arr:
        list_permutations[i] += 1
    count_perm = list_permutations[1]
    for i in range(1, max_elem):
        if list_permutations[i] < list_permutations[i+1]:
            return -1, []
    for i in range(len(arr)):
        num = arr[i]
        arr[i] = list_permutations[num]
        list_permutations[num] -= 1
    return count_perm, arr

def check_task1(n: int, arr: list[int]):
    check = True
    if n not in range(1,106):
        check = False
        return check
    for i in arr:
        if i not in range(1,106):
            check = False
            break
    return check

arr = []
try:
    n = int(input("Количество целых чисел в перестановке: "))
    while len(arr)!=n:
        print("Введите", n, "положительных чисел")
        arr = [int(x) for x in input().split()]       
except ValueError:
    arr = []
    n = 0

if check_task1(n, arr):
    res = permutation(arr)
    print("Количество перестановок:", res[0])
    if res[1]!=[]:
        print("Перестановки по группам: ")
        for i in res[1]:
            print(i, end=' ')
else:
    print("Введены некоректные данные")
```

## Задача 3
Вася изучает позиционные системы счисления. К сожалению, он часто забывает записать основание системы счисления, в которой записано выражение. Однажды он увидел в своей тетради запись a + b = ?, при этом основание системы счисления нигде записано не было. Теперь Вася должен выбрать некоторое основание p и считать что выражение записано в p-ичной системе счисления. Вася понял, что при разных основаниях он может получать разные ответы, а некоторые основания даже недопустимы. Например, выражение 78 + 87 в 16-ичной системе счисления равняется FF16, в 15-ичной: 11015, в 10-ичной: 16510, в 9-ичной: 1769, а в системах счисления с основанием 8 или меньше данное выражение недопустимо, т. к. все цифры должны быть строго меньше основания системы счисления. Васе стало интересно, какова длина самого длинного возможного значения выражения. Помогите ему найти эту длину.

Под длиной числа следует понимать количество цифр в нем. Например, длина самого длинного ответа для 78 + 87 = ? — это 3. Это получается, например, в 15-ичной (11015), 10-ичной (16510), 9-ичной (1769) и некоторых других системах счисления.

Входные данные
В первой строке через пробел записано два числа a и b (1 ≤ a, b ≤ 1000) — заданные слагаемые.

Выходные данные
Выведите одно число — длину наидлиннейшего ответа.
Примеры\
входные данные\
78 87\
выходные данные\
3

```python
def search_number_system(num1: int, num2: int):
    num_sys = 0
    num1_str = [int(i) for i in str(num1)]
    num2_str = [int(i) for i in str(num2)]
    max1 = max(num1_str)
    max2 = max(num2_str)
    num_sys = max(max1, max2)+1
    res = []
    while len(num1_str)>len(num2_str):
        num2_str.insert(0,0)
    while len(num1_str)<len(num2_str):
        num1_str.insert(0,0)
    for i in range(len(num1_str) - 1, -1, -1):
        summ = num1_str[i] + num2_str[i]
        if summ >= num_sys:
            if i == 0:
                res.insert(0, summ-num_sys)
                res.insert(0, summ//num_sys)
            else:
                res.insert(0, summ-num_sys)
                num1_str[i-1] += summ//num_sys
        else:
            res.insert(0, summ)
    return len(res)

def check_task3(num1: int, num2: int):
    check = True
    if num1 not in range(1,1001) or num2 not in range(1,1001):
        check = False
    return check

try:
    num1, num2 = [int(x) for x in input("Введите числа: ").split()]
except ValueError:
    num1 = 0
    num2 = 0
    
if check_task3(num1, num2):
    print("Максимальная длина числа:", search_number_system(num1,num2))
else:
    print("Введены некоректные данные")
```

## Задача 4
Красная шапочка нарисовала n прямоугольников. Углы прямоугольников имеют целочисленные координаты, а их ребра параллельны осям Ox и Oy. Прямоугольники могут касаться друг друга, но не могут пересекаться (то есть не существует точки, которая принадлежит внутренности более одного прямоугольника).

Ваша задача — определить, образуют ли нарисованные прямоугольники квадрат. Иными словами, определите, существует ли такой квадрат, что множество точек, лежащих внутри или на границе хотя бы одного прямоугольника, в точности равно множеству точек, лежащих внутри или на границе этого квадрата.

Входные данные
Первая строка содержит целое число n (1 ≤ n ≤ 5). Каждая из следующих n строк содержит четыре целых числа, которые описывают один прямоугольник: x1, y1, x2, y2 (0 ≤ x1 < x2 ≤ 31400, 0 ≤ y1 < y2 ≤ 31400) — x1 и x2 обозначают x-координаты левой и правой стороны прямоугольника, y1 и y2 обозначают y-координаты нижней и верхней стороны прямоугольника.

Прямоугольники не пересекаются (то есть не существует точки, которая принадлежит внутренности более одного прямоугольника).

Выходные данные
В единственную строку выведите «YES», если нарисованные прямоугольники образуют квадрат, или «NO» в противном случае.
Примеры\
входные данные\
5\
0 0 2 3\
0 3 3 5\
2 0 5 2\
3 2 5 5\
2 2 3 3\
выходные данные\
YES

```python
def is_square(coords: list[int]):
    square = 0
    min_x = 0
    min_y = 0
    max_x = 0
    max_y = 0
    for rectangle in coords:
        square += (rectangle[2]-rectangle[0])*(rectangle[3]-rectangle[1])
        if min_x > rectangle[0]:
            min_x = rectangle[0]
        if min_y > rectangle[1]:
            min_y = rectangle[1]
        if max_x < rectangle[2]:
            max_x = rectangle[2]
        if max_y < rectangle[3]:
            max_y = rectangle[3]
    if square == (max_x - min_x)*(max_y - min_y) and max_x - min_x == max_y - min_y:
        return "YES"
    else:
        return "NO"

def check_task4(n: int, coords: list[int]):
    check = True
    if n not in range(1,6):
        check = False
        return check
    for coord in coords:
        if len(coord) != 4:
            check = False
            break
        if coord[0] not in range(0,31400):
            check = False
            break
        if coord[1] not in range(0,31400):
            check = False
            break
        if coord[2] not in range(0,31400) or coord[2] <= coord[0]:
            check = False
            break
        if coord[3] not in range(0,31400) or coord[3] <= coord[1]:
            check = False
            break
    return check

coords = []
try:
    n = int(input("Количество прямоугольников: "))
    print("Координаты прямоугольников:")
    for i in range(n):
        arr = [int(x) for x in input().split()]
        coords.append(arr)
except ValueError:
    n = 0
    coords = []   
if check_task4(n, coords):
    print("Образует квадрат:", is_square(coords))
else:
    print("Введены некоректные данные")
```

```python
def table(matr: list[int], count_list: list[int]):
    coords = [0]*(sum(count_list)+1)
    res = []
    for i in range(len(count_list)):
        for j in range(len(matr[i])):
            coords[matr[i][j]] = (i+1, j+1)
    
    k = 1
    for i in range(len(count_list)):
        for j in range(len(matr[i])):           
            if coords[k] != (i+1, j+1):            
                value = coords.index((i+1, j+1))
                exchange_coords = coords[k]
                res.append(str(i+1) + ' ' + str(j+1) + ' ' + str(exchange_coords[0]) + ' ' + str(exchange_coords[1]))
                coords[value] = exchange_coords
                coords[k] = (i+1, j+1)
            k += 1
    return res

def check_task5(n: int, matr: list[int], count_list: list[int]):
    check = True
    if n not in range(1,51):
        check = False
        return check
    if len(count_list)!=n or n != len(matr):
        check = False
        return check
    for i in range(n-1):
        if count_list[i] not in range(1,51) or count_list[i] < count_list[i+1]:
            check = False
            break
    if count_list[n-1] not in range(1,51):
        check = False
    s = sum(count_list)
    count_aij = [0]*(s+1)
    for i in range(n):
        if count_list[i] != len(matr[i]):
            check = False
            break
        for j in range(len(matr[i])):
            if matr[i][j] not in range(1,s+1):
                check = False
                break
            count_aij[matr[i][j]] += 1
            
    for i in count_aij:
        if i > 1:
            check = False
            break
    
    return check

matr = []
try:
    n = int(input("Количество строк: ")) 
    count_list = [int(x) for x in input("Количество клеточек в строках: ").split()]
    print("Таблица: ")
    for i in range(n):
        matr.append([int(x) for x in input().split()])
except ValueError:
    n = 0
    matr = []
    count_list = []
if check_task5(n, matr, count_list):
    res = table(matr, count_list)
    print("Количество перестановок:",len(res))
    print("Координаты перестановок:")
    for i in res:
        print(i)
else:
    print("Введены некоректные данные")
```

И тогда перед Петей встал весьма нетривиальный вопрос: какую букву удалять?

Входные данные
Входные данные содержат две строки, состоящие из строчных латинских букв. Длина строк от 1 до 106 символов включительно, первая строка содержит ровно на 1 символ больше, чем вторая.

Выходные данные

Во второй строке через пробел выведите сами позиции в порядке возрастания. Позиции нумеруются с 1. Если из первой строки невозможно получить вторую путем удаления одного символа, выведите одно число 0.

```python
def delete_letter(wrong_word: str, correct_word: str):
    count = 0
    position = []
    for i in range(len(wrong_word)-1):
        if correct_word == wrong_word[:i] + wrong_word[i+1:]:
            count += 1
            position.append(i+1)
    return count, position

def check_task6(wrong_word: str, correct_word: str):
    check = True
    if len(wrong_word)-1 != len(correct_word) or len(correct_word) not in range(1,106):
        check = False
        return check
    for i in range(len(correct_word)):
        if ord(correct_word[i])<97 or ord(correct_word[i])>122:
            check = False
            break
        if ord(wrong_word[i])<97 or ord(wrong_word[i])>122:
            check = False
            break
    if ord(wrong_word[-1])<97 or ord(wrong_word[-1])>122:
        check = False
        
    return check 
    
wrong_word = input("Слово с ошибкой: ")
correct_word = input("Правильное слово: ")
if check_task6(wrong_word, correct_word):
    res = delete_letter(wrong_word, correct_word)
    print("Количество позиций:",res[0])
    print("Позиции:", end = ' ')
    if res[0] != 0:
        for i in res[1]:
            print(i, end = ' ')
else:
    print("Введены некоректные данные")
```

Во-первых, типом является строка «int».

Гена с радостью помог Васе, он продиктовал ему тип языка X--, который хранит n целых чисел. К сожалению, Гена торопился и не произносил знаков препинания. Теперь Гена уже ушел, а Вася все никак не может расставить знаки препинания, чтобы получить тип языка X--.

Выходные данные

Гарантируется, что если такой тип существует, то он единственен.

Обратите внимание, что нужно вывести тип, произнесенный Геной (если такой существует), а не любой, в котором помещается n значений.

```python
def correct_language(code: list[str]):
    res = ""
    count_pair = 0
    count_int = 0
    count_left = 0
    count_right = 0
    for i in range(len(code)):
        if len(code)!=1 and code[0]=="int":
            return "Error occurred"
        elif len(code)==1 and code[0]=="int":
            res = "int"
            return res
        if code[i] == "pair":
            res += code[i] + "<"
            count_left += 1
            count_pair += 1
            continue
        elif i > 0 and code[i] == "int" and code[i-1] == "pair":
            res += code[i] + ","
            count_int += 1
            continue
        elif i > 0 and code[i] == "int" and res[-1]==",":
            res += code[i] + ">"
            count_right += 1
            count_int += 1
            continue
        elif i > 0 and code[i] == "int" and res[-1]==">":
            res += "," + code[i] + ">"
            count_right += 1
            count_int += 1
            continue    
    while res[-1]==">" and count_int-1 == count_pair and count_left != count_right:
        res += ">"
        count_right += 1
    if count_left == count_right and res[-1]==">":
        return res
    else:
        return "Error occurred"

def check_task7(n: int, code: list[str]):
    check = True
    count = 0
    if n not in range(1,106):
        check = False
        return check
    for command in code:
        if command != "pair" and command != "int":
            check = False
            break
        if command == "int":
            count += 1
    if n != count:
        check = False       
    return check
try:
    n = int(input("Количество int: "))
    code = input("Код: ").split()
except ValueError:
    n = 0
    code = []
if check_task7(n,code):
    print("Верный код:",correct_language(code))
else:
    print("Введены некоректные данные")
```

```python
def count_course(k: list[int], program: list[int]):
    count = 0
    course = []
    for i in k:
        if i > len(program):
            return -1
        index_history = []
        index_history.append(i-1)
        tree = tree_of_top_course(program, i-1, index_history)
        course.extend(tree)
        if -1 in course:
            return -1
        count += len(tree)
    for i in course: 
        if course.count(i) > 1: 
            course.remove(i)
    return count, course
        
def tree_of_top_course(program: list[int], index: int, index_history: list[int]):
    tree = []
    if program[index][0] == 0:
        tree.append(index + 1)
        return tree
    for i in range(1, len(program[index])):
        if program[index][i]-1 in index_history:
            tree.append(-1)
            return tree
        index_history.append(program[index][i]-1)
        tree.extend(tree_of_top_course(program, program[index][i]-1, index_history))
    tree.append(index + 1)
    return tree

def check_task8(n: list[int], k: list[int], program: list[int]):
    check = True
    if len(n) != 2 or n[0] not in range(1,106) or n[1] not in range(1,106) or n[0]<n[1]:
        check = False
        return check
    if len(k)!=n[1]:
        check = False
        return check
    for i in k:
        if i not in range(1,n[0]+1):
            check = False
            break
    for i in range(len(program)):
        if program[i][0] != len(program[i])-1 or program[i][0] not in range(0,n[0]):
            check = False
            break
        for j in range(1, len(program[i])):
            if program[i][j] not in range(1,n[0]+1) or program[i][j] == i+1:
                check = False
                break
    return check

program = []
try:
    n = [int(x) for x in input("Количество курсов и количество главных курсов: ").split()]
    k = [int(x) for x in input("Главные курсы: ").split()]
    print("Информация о курсах:")
    for i in range(n[0]):
        program.append([int(x) for x in input().split()])
except ValueError:
    n = []
    k = []
    program = []
if check_task8(n, k, program):
    res = count_course(k, program)
    if type(res) is tuple:
        print("Количество курсов для получения специальности:", res[0])
        print("Курсы в порядке прохождения:", end = " ")
        for i in res[1]:
            print(i, end = " ")
    else:
        print("Невозможно составить план курсов:", res)
else:
    print("Введены некоректные данные")
```

Из пункта A в пункт B со скоростью v м/с едет машина. Действие происходит на оси Ox. На расстоянии d метров, считая от A, находится светофор. Начиная с момента времени 0, первые g секунд на светофоре говорит зеленый свет, следующие r секунд на нем говорит красный свет, затем снова g секунд зеленый и так далее.

Машина может мгновенно разгоняться от 0 до v и, наоборот, может мгновенно тормозить от v до 0. Считается, что она проезжает на светофоре на зеленый свет мгновенно. Если машина подъезжает к светофору в момент, когда на нем только загорелся красный, она не успевает проехать. Но если она подъезжает в момент, когда только загорелся зеленый, она может проехать. Машина выезжает из пункта A в момент времени 0.

За какое наименьшее время машина сможет доехать из пункта A до пункта B, не нарушая при этом правил дорожного движения?

Входные данные
В первой строке записаны целые числа l, d, v, g, r (1 ≤ l, d, v, g, r ≤ 1000, d < l) — соответственно расстояние между A и B (в метрах), расстояние от A до светофора, скорость машины, время зеленого света, время красного света.

Выходные данные
Выведите одно число — за какое наименьшее время машина сможет приехать из пункта A в пункт B. Ваш ответ должен иметь относительную или абсолютную погрешность меньше чем 10 - 6.

Примеры\
входные данные\
2 1 3 4 5\
выходные данные\
0.66666667

```python
def average_time(l: int, d: int, v: int, g: int, r: int):
    time = 0
    time_to_light = d/v
    if time_to_light % (g + r) < g:
        time = time_to_light + ((l - d)/v)
    else:
        time_to_light += (g + r) - (time_to_light % (g + r))
        time = time_to_light + ((l - d)/v)
    return "{0:.8f}".format(time)

def check_task9(data: list[int]):
    check = True
    if len(data) != 5:
        check = False
        return check
    if data[0] not in range(1,1001):
        check = False
        return check
    if data[1] not in range(1,1001):
        check = False
        return check
    if data[2] not in range(1,1001):
        check = False
        return check
    if data[3] not in range(1,1001):
        check = False
        return check
    if data[4] not in range(1,1001):
        check = False
        return check
    if data[1] >= data[0]:
        check = False
        return check
    return check

try:
    data = [int(x) for x in input("Введите расстояние между двумя пунктами, расстояние до светофора, скорость машины, время горения зелёного и красного светофора:").split()]
except ValueError:
    data = []
if check_task9(data):
    print("Наименьшее время:", average_time(data[0],data[1],data[2],data[3],data[4]))
else:
    print("Введены некоректные данные")
```

Учтите, что если x не нашлось, то операция засчитывается как выполненная, и массив не меняется вообще.

Вам дано исходное число (в виде массива цифр) и число k. Помогите Пете узнать результат выполнения k операций.

Входные данные
В первой строке задано два целых числа n и k (1 ≤ n ≤ 105, 0 ≤ k ≤ 109) — количество цифр числа и количество сделанных операций. Во второй строке задано n цифр без пробелов — массив цифр d, начиная с d1. Гарантируется, что первая цифра числа не равна нулю.

Выходные данные
В единственной строке выведите без пробелов результат — число после выполнения k операций.

Примеры\
входные данные\
7 4\
4727447\
выходные данные\
4427477

```python
def lucky_number(num: list[int], k: int):
    while k != 0:
        try:
            index4 = num.index(4)
            index7 = num.index(7)
        except ValueError:
            index4 = -1
        while index4 + 1 != index7:
            if index4 == -1:
                break
            if index7 < index4:
                start = index7 + 1
            else:
                start = index4 + 1
            try:
                index4 = num.index(4, start)
                index7 = num.index(7, start)
            except ValueError:
                index4 = -1
        if index4 == -1:
            k -= 1
            break
        if (index4+1) % 2 != 0:
            num[index7] = 4
        else:
            num[index4] = 7
        k -= 1
        
    return "".join(map(str, num))

def check_task10(n: int, k: int, num: str):
    check = True
    if n not in range(1, 106):
        check = False
        return check
    if k not in range(0, 110):
        check = False
        return check
    if len(num) != n:
        check = False
        return check
    for i in num:
        if ord(i)<48 or ord(i)>57:
            check = False
            break
    return check

num = []
try:
    n, k = [int(x) for x in input("Количество цифр и количество операций: ").split()]
    num_str = input("Массив цифр (без пробелов): ")
    for i in range(len(num_str)):
        num.append(int(num_str[i]))
except ValueError:
    n = 0
    k = 0
    num_str = ""
if check_task10(n, k, num_str):
    print("Результат преобразований:",lucky_number(num, k))
else:
    print("Введены некоректные данные")
```
